```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
import chromedriver_autoinstaller
from selenium import webdriver
```

```python
#INSTALL BROWSER SUCH AS CHROME
chromedriver_autoinstaller.install()

#BROWSER SELENIUM
driver = webdriver.Chrome()
```

```python
driver.get("https://mydramalist.com/shows/top_korean_dramas")
```

```python
soup = BeautifulSoup(driver.page_source,"lxml")
```

```python
soup
```

```python
kdrama = soup.find_all("div", class_="row")
```

```python
data = []
for row in kdrama :
    title = row.find(class_ ='title')
    desc = row.find(class_ ='text-muted')
    score = row.find(class_ ='score')
    
    kdramas = {
        'title' : title,
        'desc' : desc,
        'score' : score
    }
    
    data.append(kdramas)
```

```python
df=pd.DataFrame(data)
```
